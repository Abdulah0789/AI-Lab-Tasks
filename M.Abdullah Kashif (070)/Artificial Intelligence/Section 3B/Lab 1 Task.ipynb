Hangman Dynamic

```python
import random

def hangman(tries):
        stages = [
        '''
           -----
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        --------
        ''',
        '''
           -----
           |   |
           O   |
          /|\\  |
          /    |
               |
        --------
        ''',
        '''
           -----
           |   |
           O   |
          /|\\  |
               |
               |
        --------
        ''',
        '''
           -----
           |   |
           O   |
          /|   |
               |
               |
        --------
        ''',
        '''
           -----
           |   |
           O   |
           |   |
               |
               |
        --------
        ''',
        '''
           -----
           |   |
           O   |
               |
               |
               |
        --------
        ''',
        '''
           -----
           |   |
               |
               |
               |
               |
        --------
        '''
    ]
        print(stages[tries])

def hangmane():
        words = ['apple', 'banana', 'grape', 'melon']
        word = random.choice(words)
        guessed = ['_'] * len(word)
        tries = 6
        guessed_letters = []

        print("Welcome to Hangman!")

        while tries > 0 and '_' in guessed:
            hangman(tries)
            print(' '.join(guessed))
            print(f"Guessed Letters: {', '.join(guessed_letters)}")
            guess = input("Enter a letter: ").lower()

            if guess in guessed_letters:
                   print("You already guessed that letter!")
                   continue
            
            guessed_letters.append(guess)

            if guess in word:
                for i in range(len(word)):
                    if word[i] == guess:
                        guessed[i] = guess
                print(f"Good Job! '{guess}' is in the word.")
            else:
                tries -= 1
                print(f"Wrong! '{guess}' is not in the word. {tries} attempts left.")
        hangman(tries)

        if '_' not in guessed:
            print(f"Congratulations! You guessed the word: {''.join(guessed)}")
        else:
            print(f"You lost! The word was: {word}")

hangmane()
```
